# **Eksplorasi Data (EDA)**

Pada tahap ini dilakukan pemeriksaan terhadap struktur data, tipe variabel, serta gambaran umum isi dataset sebagai langkah awal sebelum dilakukan analisis lebih lanjut.

## **Comparison/Perbandingan**

Comparison atau perbandingan adalah proses membandingkan dua atau lebih data, nilai, atau kategori untuk melihat perbedaan, persamaan, atau kecenderungan di antara mereka.

```python
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x='Brand',
    order=df['Brand'].value_counts().index,
    palette='RdPu'
)

plt.title('Jumlah Penjualan Berdasarkan Brand')
plt.xticks(rotation=45)
plt.show()
```

Grafik ini menunjukkan perbandingan jumlah transaksi penjualan pada setiap brand makeup. Brand dengan jumlah transaksi tertinggi (MAC) menandakan bahwa produk dari brand tersebut lebih banyak diminati oleh konsumen dibanding brand lainnya. Perbedaan jumlah transaksi antar brand juga menunjukkan adanya persaingan performa penjualan dalam pasar kosmetik global tahun 2025.

## **Composition/Komposisi**

Composition (komposisi) merupakan teknik dalam analisis data yang digunakan untuk menunjukkan bagaimana suatu keseluruhan terbagi ke dalam beberapa bagian atau kategori. Pendekatan ini bertujuan untuk memahami proporsi atau persentase masing-masing bagian terhadap total data, sehingga dapat diketahui bagian mana yang paling dominan maupun yang memiliki kontribusi paling kecil dalam suatu dataset.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Revenue_USD'],
    bins=20,
    kde=True,
    color='#FF69B4'
)

plt.title('Distribusi Revenue_USD')
plt.show()
```

*Histogram* menunjukkan distribusi pendapatan penjualan produk makeup. Sebagian besar transaksi berada pada rentang *revenue* menengah, namun terdapat beberapa transaksi dengan *revenue* yang sangat tinggi. Pola distribusi yang cenderung *right-skewed* menunjukkan bahwa hanya sebagian kecil transaksi yang memberikan pendapatan sangat besar dibanding transaksi lainnya.

## **Distribution/Distribusi**

Distribusi merupakan teknik analisis data yang digunakan untuk melihat penyebaran nilai dari suatu variabel numerik dalam dataset. Melalui distribusi, dapat diketahui pola sebaran data, seperti kecenderungan nilai, tingkat variasi, serta adanya kemungkinan nilai ekstrem (*outlier*).

Pie chart di atas menunjukkan proporsi penggunaan setiap saluran penjualan dalam transaksi produk makeup. Saluran penjualan dengan persentase terbesar (*Online*) menunjukkan metode distribusi yang paling dominan digunakan oleh konsumen. Perbedaan proporsi ini dapat menggambarkan preferensi konsumen terhadap cara pembelian produk makeup pada tahun 2025.

## **Relationship/Hubungan**

Relationship merupakan teknik analisis data yang digunakan untuk melihat hubungan atau keterkaitan antar variabel dalam dataset. Melalui analisis relationship, dapat diketahui apakah suatu variabel memiliki pengaruh atau hubungan terhadap variabel lainnya, serta seberapa kuat hubungan tersebut terjadi. Analisis ini membantu dalam memahami pola keterkaitan data dan faktor-faktor yang dapat memengaruhi suatu variabel dalam dataset.

```python
plt.figure(figsize=(6,4))

numeric_cols = df[['Price_USD', 'Units_Sold', 'Revenue_USD']]

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='RdPu',
    linewidths=2,
    linecolor='white'
)

plt.title('Relationship Between Makeup Sales Variables in 2025')
plt.show()
```

Berdasarkan hasil heatmap correlation, terlihat bahwa `Revenue_USD` memiliki hubungan positif dengan `Price_USD` dan `Units_Sold`. Hal ini menunjukkan bahwa semakin tinggi harga produk dan semakin banyak unit yang terjual, maka pendapatan penjualan cenderung meningkat. Selain itu, hubungan antar variabel dalam dataset menunjukkan korelasi yang cukup baik dalam memengaruhi performa penjualan makeup tahun 2025.

## **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Berdasarkan hasil deteksi outlier menggunakan metode IQR, secara umum dataset tidak memiliki nilai ekstrem yang signifikan pada sebagian besar kolom numerik. Hal ini menunjukkan bahwa data memiliki distribusi yang cukup stabil dan konsisten sehingga aman digunakan untuk proses analisis.

* **Revenue_USD**

  Terdapat sekitar 1,2% data yang terdeteksi sebagai *outlier*. Namun, persentase tersebut sangat kecil sehingga tidak memberikan pengaruh besar terhadap distribusi data secara keseluruhan. Nilai tersebut masih dapat dianggap wajar karena kemungkinan berasal dari transaksi dengan total pendapatan yang lebih tinggi dibanding transaksi lainnya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Outlier pada kolom **Revenue_USD** tidak dilakukan penghapusan karena jumlahnya sangat kecil, yaitu sekitar **1,2%** dari keseluruhan data. Selain itu, nilai tersebut masih dianggap wajar dan kemungkinan berasal dari transaksi dengan pendapatan yang lebih tinggi dibanding transaksi lainnya. Oleh karena itu, outlier tetap dipertahankan agar informasi penting dalam data tidak hilang, sehingga proses analisis dapat dilanjutkan ke tahap selanjutnya.
